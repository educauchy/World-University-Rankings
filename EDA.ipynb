```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure, savefig
import seaborn as sns
```

```python
media_dir = 'Media/EDA/'
data = pd.read_excel('Data/Russian_Universities.xlsx')
data.head()

inputs = list(range(1, 53))
inputs.remove(47)
outputs = ['E1', 'E2', 'E3', 'E4', 'E5', 'E6', 'E7']
```

```python
# data.describe()
data.head()
```

```python
corr = data.iloc[:, 8:63].corr()
figure(figsize = (25, 25))
sns.heatmap(corr, annot = True, vmin = -1, vmax = 1, cmap = 'coolwarm')
plt.savefig(media_dir + 'heatmap.png')
```

```python
for i in inputs:
    if i == 47:
        continue
    figure(figsize = (20, 10))
    plt.plot(data.University, data[str(i)])
    plt.title(str(i))
    plt.xticks(rotation=90)
    plt.savefig(media_dir + str(i) + '.png')
    plt.close()
```

```python
for i in outputs:
    figure(figsize = (20, 10))
    plt.plot(data.University, data[str(i)])
    plt.title(str(i))
    plt.xticks(rotation=90)
    plt.savefig(media_dir + str(i) + '.png')
    plt.close()
```
