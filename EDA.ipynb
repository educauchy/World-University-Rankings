```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure, savefig
import seaborn as sns
```

```python
media_dir = 'Media/EDA/'
data = pd.read_excel('Data/Russian_Universities.xlsx')
data.head()

inputs = list(range(1, 53))
inputs.remove(47)
outputs = ['E.1', 'E.2', 'E.3', 'E.4', 'E.5', 'E.6', 'E.8']
```

```python
# data.describe()
data.head()
```

```python
corr = data.iloc[:, 8:63].corr()
figure(figsize = (25, 25))
sns.heatmap(corr, annot = True, vmin = -1, vmax = 1, cmap = 'coolwarm', fmt=".1f")
plt.savefig(media_dir + 'heatmap_inputs.png')
```

```python
corr = data.loc[:, outputs].corr()
figure(figsize = (25, 25))
sns.heatmap(corr, annot = True, vmin = -1, vmax = 1, cmap = 'coolwarm', fmt=".1f")
plt.savefig(media_dir + 'heatmap_outputs.png')
```

```python
for i in inputs:
    figure(figsize = (20, 10))
    plt.plot(data.University, data[str(i)])
    plt.title(str(i))
    plt.xticks(rotation=90)
    plt.savefig(media_dir + '/Distributions/' + str(i) + '.png')
    plt.close()
```
